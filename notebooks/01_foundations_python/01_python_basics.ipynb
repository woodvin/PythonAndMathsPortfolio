# 5. Writing Clean, Quant-Ready Code
Follow simple rules:
- Use meaningful variable names (`price`, `rate`, not `p`, `r`)
- Use consistent spacing
- Keep functions short
- Add docstrings to every function

```python
def compute_total(cost, tax_rate):
    """
    Calculates the total cost after tax.
    """
    return cost + cost * tax_rate

compute_total(100, 0.2)
```
